##### Show effect of matrix-vector and matrix-matrix multiplication on image data.

Machines store images in the form of a matrix of numbers. The size of this matrix depends on the number of pixels we have in any given image. Let’s say the dimensions of an image are n x m. These dimensions are basically the number of pixels in the image (height x width). A colored image is typically composed of multiple colors and almost all colors can be generated from three primary colors – red, green and blue. Hence, in the case of a colored image, there are three Matrices (or channels) – Red, Green, and Blue. Each matrix has values between 0-255 representing the intensity of the color for that pixel. Consider this picture of chelsea the cat:

```python
import numpy as np
import scipy.misc
import matplotlib.pyplot as plt
from skimage import data
img = data.chelsea()
plt.axis('off') 
plt.imshow(img)
print((img.shape))
```

This image is (300 x 451) x 3 pixels. It’s a matrix of dimension 300 x 451 x 3.
We can multiply this matrix with a random vector of length = 3 and see how the value of matrix value changes resulting in change in color of image.

```python
vec = np.random.rand(3)
img_tinted = img.dot(vec)
plt.axis('off') 
plt.imshow(img_tinted.astype('uint8'))
img_tinted.shape
```

##### Matrix-Vector Multiplication (Dot Product)

We can see that as we are applying dot product on matrix of shape 300 x 451 x 3 and vector of shape 3 x 1 resultant matrix is of shape 300 x 451 x 1 or 300 x 451.  
Also we can observe from the image that only one color is visible as other two components has vanished.

```python
mat = np.random.rand(300,451,3)
img_tinted_1 = img*mat
plt.axis('off') 
plt.imshow(img_tinted_1.astype('uint8'))
img_tinted_1.shape
```

##### Matrix-Matrix Multiplication (Hadamard Product)

We can see that as we are applying Hadamard product on matrix of shape 300 x 451 x 3 and matrix of shape 300 x 451 x 3 resultant matrix is of shape 300 x 451 x 3 and only values are changing.  
Also we can observe from the image that color intensity has changed as values of color components have changed.
